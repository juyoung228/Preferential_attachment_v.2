```python
# The script is for couting # of Journals in 'Papers.txt' file.
# The structure of the file is as in the following

'''
1	Paper ID
2	Original paper title
3	Normalized paper title
4	Paper publish year
5	Paper publish date 
6	Paper Document Object Identifier (DOI)
7	Original venue name
8	Normalized venue name
9	Journal ID mapped to venue name
10	Conference series ID mapped to venue name
11	Paper rank
'''

freq_map = {}
n  = 0
with open("G:\\Papers\\Papers.txt", encoding= 'utf-8') as infile:
    for line in infile:
        n += 1
        j_id = line.split("\t")[8]
        if j_id != "": 
            if j_id not in freq_map:
                freq_map[j_id] = 0
        if n%1000000 == 0:
            print(n)

    print(len(freq_map))

# result: 
# paper:122000000
# of journals 22872
```
